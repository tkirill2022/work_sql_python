# Predicting the cost of medical insurance



### Loading the required libraries

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

## Loading data:

Data collected:
- age
- gender
- bmi: body mass index
- children
- smoker
- region:(northeast, southeast, southwest, northwest).
- charges: individual medical expenses (we want to predict it)

```python
#Load data from  file 'insurance.csv' into df variable
df = pd.read_csv('insurance.csv', sep = ',')
```

```python
df.head()
```

## We look at the statistics that there are no spaces and missing values

```python
df.info()
```

```python
df.describe()
```

## Converting string data

```python
# Replacing gender and smoking with numbers
df['sex']=df['sex'].map({'male':1, 'female':0})
df['smoker']=df['smoker'].map({'yes':1,'no':0})
```

```python
df.head()
```

```python
# Replacing the region with a set of individual columns (is_southwest, is_southeast и тп)
df = pd.get_dummies(df, columns=['region'])
df.head()
```

## We form features and a target variable

```python
## Shape Features and Target Variable

df.columns
```

```python
x = df[['age', 'sex', 'bmi', 'children', 'smoker', 'region_northeast', 'region_northwest', 'region_southeast','region_southwest']]
y = df['charges']
```

```python
y
```

## Divide the data into a sample for training / validation

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
X_train
```

## Creating and training a linear regression model

```python
# Create and train a Model
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

## We get a prediction and evaluate the quality

```python
# Get a prediction
lr.predict(X_test)
```

```python
mean_squared_error(y_test, lr.predict(X_test))
```

```python
mean_squared_error(y_train, lr.predict(X_train))
```
