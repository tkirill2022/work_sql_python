```python
import pandas as pd
df_holiday = pd.read_csv('Holiday_planning.csv', sep=',')
df_city = pd.read_sql(
    """select name, population from city where country_code not in ('USA', 'VEN')""",
    'postgresql://netology:NetoSQL2019@84.201.153.170:19001/world-db'
)
```

```python
#DataFrame with city and population
df_city.head()
```

```python
#DataFrame with data from the table "Holiday planning"
df_holiday.head()
```

```python
#joining tables
df_merged = df_city.merge(df_holiday, left_on = 'name', right_on = 'City_english', how = 'inner')
```

```python
df_merged.head(15)
```

```python
df_merged.columns
```

```python
#create new dataframe with table (Country, name, population)
df_population = df_merged[['Country', 'name', 'population']]
```

```python
df_population.head(33)
```

```python
#creation of a new variable with data grouping by countries and cities
y = df_population.groupby(['Country', 'name']).mean()
```
